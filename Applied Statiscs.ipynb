# Descriptive Statistics

```python
import numpy as np 
```

```python
data = np.random.normal(100,15,1000)
```

```python
mean = np.mean(data)
median = np.median(data)
mode = float(max(set(data), key=list(data).count))

print(f"Mean: {mean:2f}")
print(f"Median:{median:.2f}")
print(f"Mode: {mode:.2f}")
```

# Measures of Variability: Standard Deviation and Variance

```python
std_dev = np.std(data)
variance = np.var(data)
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Variance:{variance:.2f}")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(range(len(data)), data, alpha=0.5)
plt.axhline(y=np.mean(data), color = 'r', linestyle = '--', label = 'Mean')
plt.axhline(y=np.mean(data) + std_dev, color = 'g', linestyle ='--', label= 'Mean + 1 Std Dev')
plt.axhline(y=np.mean(data) - std_dev, color = 'g', linestyle ='--', label= 'Mean - 1 Std Dev')
plt.legend()
plt.title('Data Dispersion Vosualization')
plt.xlabel('Data Point Index')
plt.ylabel('Value')
plt.show()
```

# Probability Distributions: Normal Distribution

```python
from scipy import stats
```

```python
mu, sigma = 0, 1
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y)
plt.title('Standard Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.fill_between(x,y, where=(x >= -sigma) & ( x <= sigma), alpha=0.3, label = "68% of data")
plt.legend()
plt.show()
```

```python
prob_within_1_sigma = stats.norm.cdf(1) - stats.norm.cdf(-1)
print(f"Probability within 1 standard deviation: {prob_within_1_sigma:.4f}")
```

# Hypothesis Testing: t-test

```python
group_1 = np.random.normal(100, 15, 50)
group_2 = np.random.normal(100, 15, 50)
```

```python
t_statistics, p_value = stats.ttest_ind(group_1, group_2)
print(f"T-statistics: {t_statistics:.4f}")
print(f"P=value: {p_value:.4f}")
```

```python
alpha = 0.5
if p_value < alpha:
    print("Reject null hypothesis: There is a significant difference between the groups.")
else:
    print("Fail to reject null hypothesis: There is no significant dofference between the groups.")
```

# Correlation and Regression Analysis

```python
x = np.random.normal(0,1,100)
y = 2* x + np.random.normal(0, 0.5, 100)
```

```python
correlation_coeff, _ = stats.pearsonr(x,y)
```

```python
sloppe, intercept, r_value, _, _ =  stats.linregress(x,y)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x, y, alpha = 0.5)
plt.plot(x, sloppe * x * intercept, color='r', label='Regression Line')
plt.title(f"Correlation and Regression Analysis (r = {correlation_coeff:.2f})")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
print(f"Correlation coefficient: {correlation_coeff:.4f}")
print(f"Regression equation: Y = {sloppe:.2f}X + {intercept:.2f}")
```
